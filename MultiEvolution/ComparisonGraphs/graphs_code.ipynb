```python
#DATA WAS COMBINED AND CLEANED USING EXCEL BEFOREHAND

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('new_combined_data.csv', sep=';', decimal=',')
df.head()
```

```python
#converting all to float
df['mut'] = df['mut'].astype(float)
df['environment'] = df['environment'].astype(str)
df['gen'] = df['gen'].astype(float)
df['mean'] = df['mean'].astype(float)
df['max'] = df['max'].astype(float)

df.head()
```

```python
df['mut']
```

```python
#splitting data into different mutations
df1 = df[(df.mut == 0.2)]
df2 = df[(df.mut == 0.3)]
df3 = df[(df.mut == 0.5)]
df4 = df[(df.mut == 0.7)]
df5 = df[(df.mut == 0.9)]
```

```python
#mutation 1

by_label = df1.groupby('environment')

#loop divides plotting data by environment
for env, group in by_label:
    #plotting lines
    plt.errorbar(group['gen'], group['mean'], yerr= group['mean'].std(), label='MEAN env '+env)
    plt.errorbar(group['gen'], group['max'], yerr= group['max'].std(), label='MAX env '+env)  
    
#graph editing
plt.legend()
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Mutation 1 (0.2)')
plt.show
```
